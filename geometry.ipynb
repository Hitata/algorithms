```python
def find_middle(point1, point2):
    """
    Given two points in 2D space, return the middle point.
    """
    x1, y1 = point1
    x2, y2 = point2
    return ((x1 + x2) / 2, (y1 + y2) / 2)
```

```python
import matplotlib.pyplot as plt

# define the x and y coordinates of the points
x = [0, 0, 0.5, 1, 1, 1, 1, 0.5, 0.5, 0.25, 0, 0, 0]
y = [0, 0.5, 1, 1, 0.5, 0, 0.5, 0.5, 0.25, 0, 0, 0.5, 1]

# plot the points
plt.plot(x, y)

# set the axis limits
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.1)

# set the size of the figure
plt.gcf().set_size_inches(100, 100)

# show the plot
plt.show()
```
