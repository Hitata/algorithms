```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(start = 0, stop = 3, num = 51)
xx = np.arange(0, 3, 0.1)
def f(x):
    return x**2*np.exp(-x**2)
def g(x):
    return x*np.exp(-x)

y = f(x)
yy = g(xx)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(xx, yy)
plt.show()
```

```python
import matplotlib.pyplot as plt

# create a hexagon
hexagon = plt.Polygon([(0, 1), (1, 2), (2, 1), (2, 0), (1, -1), (0, 0)], closed=True)

# plot the hexagon
fig, ax = plt.subplots()
ax.add_patch(hexagon)
ax.set_xlim(-1, 3)
ax.set_ylim(-2, 2)
plt.show()

```
